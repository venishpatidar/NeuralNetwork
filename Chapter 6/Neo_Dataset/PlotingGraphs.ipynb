```python
import numpy as np
import matplotlib.pyplot as plt
import csv
from sklearn import metrics
def plot_loss(lst_iter, slt_loss, title):
    plt.plot(lst_iter, slt_loss, 'b-', label='loss')
    plt.xlabel("n iteration")
    plt.legend(loc='upper right')
    plt.title(title)
    plt.ylabel("losses")
    plt.show()
```

## Plotting Loss

```python
NEO_loss = []
with open("NEO_losses.csv") as f:
    csv_reader = csv.reader(f)
    for row in csv_reader:
        NEO_loss.append(float(row[0]))
NEO_loss = np.array(NEO_loss)
plot_loss(range(len(NEO_loss)),NEO_loss,"NEO loss")
```

## Plotting the confusion matrix

```python
actual = []
with open("Y_TEST_NEO.csv") as f:
    csv_reader = csv.reader(f)
    for row in csv_reader:
        actual.append(float(row[0]))
actual = np.array(actual)
```

```python
NEO_Predicted = []
with open("Y_PREDICTS.csv") as f:
    csv_reader = csv.reader(f)
    for row in csv_reader:
        NEO_Predicted.append(float(row[0]))
NEO_Predicted = np.array(NEO_Predicted)
```

```python
confusion_matrix = metrics.confusion_matrix(actual, NEO_Predicted)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])
cm_display.plot()
plt.title("NEO confusion matrix")
plt.show()
```
